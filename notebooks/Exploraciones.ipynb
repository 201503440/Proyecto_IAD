```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Definir el directorio en donde estan los archivos CSV
directorio = '../data'

# Lista para almacenar los DataFrames
dataframes = []

# Itera a los archivos en el directorio
for archivo in os.listdir(directorio):
    if archivo.endswith('.csv'):
        df = pd.read_csv(os.path.join(directorio, archivo))
        dataframes.append(df)

print(f'Se han cargado {len(dataframes)} archivos CSV.')
```

```python
# Ver las primeras filas de cada datasframe
for i, df in enumerate(dataframes):
    print(f'Archivo {i + 1}:')
    print(df.head(), '\n')
```

```python
# Verificar la informacion de cada dataframe
for i, df in enumerate(dataframes):
    print(f'Informacion del archivo {i + 1}:')
    print(df.info(), '\n')
```
